```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
size = 1
n = 2**(size + 2**size)

def create_mux(x):
    size = np.int(np.log2(x.shape[1]))
    a = x[:, :2**size]
    b = x[:, -size:]
    c = np.sum([(b[:, size-i-1]>0.5) * 2**i for i in range(size)], axis=0)  # Selected

    return np.array([1/a[i, 2**size-int(c[i])-1] for i in range(a.shape[0])])

# x = np.random.randint(0, 2, size=(n, 2**size+size))  # Random bits
# x = np.array([[int(j) for j in np.binary_repr(i, width=size + 2**size)] for i in range(n-1)])  # First seven

#x = np.concatenate((np.random.rand(200, 2), np.random.randint(0, 2, size=(200, 1))), axis=1)
x = (np.random.rand(200, 3)+0.2)/1.2
print(x)
y = create_mux(x)
np.save("simple.npy", np.concatenate((x,y.reshape(-1,1)), axis=1))
```
